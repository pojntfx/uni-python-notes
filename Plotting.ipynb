```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
X = np.arange(0, 2*np.pi, 0.1)
Y = np.sin(X)
Z = np.cos(X)
```

```python
plt.plot(X, Y, label="sin(x)")
plt.plot(X, Z, label="cos(x)")
plt.legend()
```
